```python
import pandas as pd
import numpy as np
import json
from glob import glob
```

```python
path = glob('data/*.json')
```

```python
len(path)
```

```python
labels_dict = dict(filename = [], text = [], xmin = [], xmax = [], ymin = [], ymax = [])
```

```python
for filepath in path:
    file = open(filepath)
    info = json.load(file)
    if len(info['results']) > 0:
        xmin = int(info['results'][0]['box']['xmin'])
        xmax = int(info['results'][0]['box']['xmax'])
        ymin = int(info['results'][0]['box']['ymin'])
        ymax = int(info['results'][0]['box']['ymax'])
        filename = info['filename']
        text = info['results'][0]['plate'].upper()


        labels_dict['filename'].append(filename)
        labels_dict['text'].append(text)
        labels_dict['xmin'].append(xmin)
        labels_dict['xmax'].append(xmax)
        labels_dict['ymin'].append(ymin)
        labels_dict['ymax'].append(ymax)
```

```python
raw_df = pd.DataFrame(labels_dict)
```

```python
import os, sys
from PIL import Image
import cv2
import pandas as pd
```

```python
ANNOTATION_FILE = 'labels.csv'
df = pd.read_csv(ANNOTATION_FILE)
IMG_DIR = 'data'
CROP_DIR = 'crop_dir'

files = df['filename']
size = (200,200)

for file in files:
	print(file)
	img = cv2.imread(IMG_DIR +'/' + file)
	annot_data = df[df['filename'] == file]; annot_data.drop_duplicates();
	xmin = int(annot_data['xmin'])
	ymin = int(annot_data['ymin'])
	xmax = int(annot_data['xmax'])
	ymax = int(annot_data['ymax'])
	crop = img[ymin:ymax,xmin:xmax]
	new_crop = cv2.resize(crop, dsize=size, interpolation=cv2.INTER_CUBIC)
	# new_crop.save(CROP_DIR + '/' + file.split('.')[0] + '.png', 'PNG', quality=90)
	cv2.imwrite(CROP_DIR + '/' + file.split('.')[0] + '.png', new_crop)
```
